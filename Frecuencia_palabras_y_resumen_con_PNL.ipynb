```python
#Importación de librerías
import nltk
import numpy as np
import matplotlib.pyplot as plt 
import unicodedata
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from nltk import FreqDist
nltk.download("stopwords")
nltk.download('punkt')
```

```python
#Cargamos el texto
documento = open('../PROCESAMIENTO_LENGUAJE_NATUAL/procesamiento_lenguaje_natural.txt', encoding='utf-8')
texto = documento.read()
print(texto)
```

```python
#Definimos función para remover acentos
def quitar_acentos(input_str):
    nfkd_form = unicodedata.normalize('NFKD', input_str)
    return ''.join([c for c in nfkd_form if not unicodedata.combining(c)])
```

```python
#Tokenizar
tok = word_tokenize(texto)

#Quitamos los stopwords
stop_words = set(stopwords.words('spanish'))
filtered_list = []

for word in tok:
    word = quitar_acentos(word)  # Elimina acentos
    if word.casefold() not in stop_words and word.isalnum():
        filtered_list.append(word)

filtered_list
```

```python
#Observamos las frecuencias
freq =  FreqDist(filtered_list)
names = list(freq.keys())
values = list(freq.values())

plt.figure(figsize=(30, 20))

plt.bar(range(len(freq)), values, tick_label=names)
plt.title('Frecuencias de las palabras')
plt.xticks(rotation=90)

plt.show()  
```

```python

# Obtener las 10 palabras más frecuentes
top_10 = freq.most_common(10)
top_10, top_10_frecuencias = zip(*top_10)

# Crear un gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(range(len(top_10)), top_10_frecuencias, tick_label=top_10)
plt.title('Top 10 Frecuencias más Altas')
plt.xlabel('Palabras')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45)  # Rotar etiquetas del eje x para mejor legibilidad

# Mostrar el gráfico
plt.show()
```

```python
#Resumen:
resumen = " ".join(top_10)
resumen
```
